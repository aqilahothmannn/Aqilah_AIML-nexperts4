# $$Loops$$

- A loop is a part of code meant for repeating a block of code certain number of times
- This removes the need of writing a block of code a gazillion times

- When we needed to repeat a line of code many times

- In C: 

```C
for(i=0, i<5, i++){
    print(idrees)
}
```

Two types of loops:
1. For ---> repeat for each item in a sequence or range of numbers
2. While ---> repeat while a condition is true

- Syntax for loop in Python:
```python
# For Loop
for <integer_variable> in <sequence>:
    # block of code

# while loop
while <condition>:
    # block of code

```


- it is very common to use `range()` as the sequence for the loop

````python
print("idrees is awesome")
print("idrees is awesome")
print("idrees is awesome")
print("idrees is awesome")
print("idrees is awesome")
print("idrees is awesome")
````

````python
for i in range(5): # Five elements --- 0 to 4
    print(i)
````

````python
for i in range(0, 10, 3): # start, end, step
    print(i)
````

````python
for i in [21,22,23,24,25,26]: print("idrees is awesome", i) 
````

````python
21 in [21,22,23,24,25,26]
````

````python
count_flag = 0

while count_flag < 10:
    print("idrees", count_flag)

    # count_flag = count_flag + 1
    count_flag += 1  # compound operator
````

````python
print("table generator --------------")
while True: # infinite loop running in python
    
    action = input("Choose an Action: 1 to calculate --- 2 or 'quit' or 'q for exit")

    if action in ['2', 'quit', 'q']:
        break

    if action == '1':
        num = int(input("Choose a Number: "))
        print(f"-------Table of {num} --------")
        for i in range(1,11):
            print(num, "x", i, "=", num*i) 
````

````python
line = ""

for i in range(5):
    # print("")
    for j in range(i):
        line = line + "*"
    print(line)
        
````

````python
for row in range(1,10):
    line = ""
    for col in range(10 - row):
        line = line + "*"
    print(line)
````
