### Logistic Regression
- it uses the logistic function ---> sigmoid function
- function is veryyyy important --- In DEEP LEARNING too
- 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
   return 1 / (1 + np.e **-z) # e raised to power -z
   # 1 / (1 + np.exp(-z))
```

```python
np.e # 2.718281828459045
np.exp(1) # 2.718281828459045 ** 1
np.exp(2)
```

```python
# Create a plot of the sigmoid function
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y, color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
```

- Its output is between 1 and 0
- what else in Maths is between 0 to 1????????????????????????????
    - Probability - chance from 0 to 1
    - thats why sigmoid can be used to show or plot probabilities
- Logistic Regression can give probability or Likelihood of one of two classess
