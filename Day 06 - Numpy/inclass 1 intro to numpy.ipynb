### Why do we need them

In a list, you can store multiple types of data but we cannot do that in arrays
- Storing sequential data ---> one after the other
- 
- Python does not have a built-in `array` type like some other languages
    - `array is a linear data structure that saves data sequentially`
- similar functionality and working in python can be achieved using:
    1. **Lists** 
        - Lists are a `flexible` and `commonly used` data structure for storing elements in sequence
        - allow mixing data types - `heterogenous`
        - other languages only have arrays, and they do not allow mixing types
    2. **Array module** - 
        - does allow creation of arrays
        - these are faster than lists as well
        - you have to remember stuff like `i` represents `int`
    3. **NumPy arrays** -
        - built with Numpy library

````python
import array as arr

a = arr.array('i', [1, 2, 3])

# in array package, you have to declare what type of data you are saving before hand
# 'i' represents integer type
````

#### Array 
- a data structure to save elements sequentially 
- properties:
    - saved in `contigous` memory locations (one after the other)
    - `HOMOGENOUS` - Can save only a single type of data (unlike lists)
    - faster and more memory efficient than lists
- 
- we gain massive performance instead of having flexibility
- 

#### Arrays VS Lists
- faster and more memory efficient than lists
- 
- In Lists 
    - you have a `python object` and it has its own memory overhead
    - Lists store memory addresses (pointers)
    - When accessing an element, follow pointers that are `scattered` across the RAM
    - This means `more memory used` and `less speed` in getting data in comparison to Homogenous structures like Arrays
- 
- In arrays:
    - You store `actual data`
    - No following the pointers
    - no extra memory overhead, because you exactly know how much space to allocate before time - Pack data tightly
    - Superior performance in comparison
    - 
    - Broadcasting can be done here ---> because of which Matrix Multiplication (`a @ b`)  is possible
        - a very important operation because we can solve equations because of it

#### Operations of Numpy Arrays

````python
x = np.array([1, 2, 3])
y = np.array([10, 20, 30])

# element wise operation --- does not work on lists
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x // y)
print(x ** y)
````

- Broadcasting

````python
x = np.array([1, 2, 3])

a = np.array([30])
a * x

# b = np.array([30, 10]) # (1,2) ---> (1,3)--- (2,) (3,)
b = np.array([30, 10, 10]) # (1,3) ---> (1,3)--- (3,) (3,)
b * x


x = np.array([[1],
              [2],
              [3],
              [4]])
c = np.array([
    [10, 20, 30],
    [10, 10, 10],
    [20, 30, 20],
    [10, 10, 10]
    ])

x * c # element wise
# (4,1) x (4,3)
````

##### Filtering
    - boolean masking

````python
# boolean masking
marks = np.array([34,55,65,45,69,89,76,33])

# takes list and compares every element
marks_above50_bool = marks > 50
marks_above50_bool # is the list with Boolean Values


marks_above50_actual = marks[marks_above50_bool]
marks_above50_actual


marks_above50 = marks[marks > 50]
marks_above50
````
