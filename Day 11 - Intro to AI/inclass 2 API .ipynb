# $$API$$

$$Application Programming Interface$$

- a set of rules (protocols) that allows different software systems to communicate
- 
- acts as a bridge between applications, enabling them to exchange data and functionality without exposing internal code
-  For Example:
    - a weather app uses an API to fetch live weather data from a remote server instead of building its own weather system.
    - 

How it works:

`Client Request `– The application (client) sends a request to an API endpoint (URL)

`Processing` – The API forwards the request to the server, which processes it

Response – The server sends back the requested data or an error message.

Delivery – The API returns the response to the client, often in JSON or XML format.

Common API Architectures

REST – Uses HTTP methods (GET, POST, PUT, DELETE), stateless, widely adopted.

SOAP – XML-based, strict structure, often used in enterprise systems.

GraphQL – Lets clients request exactly the data they need from a single endpoint.

gRPC – High-performance, binary protocol using Protocol Buffers.

Webhooks – Event-driven callbacks triggered by specific actions.

Types of APIs

Public (Open) – Available to any developer (e.g., Twitter API).

Private (Internal) – Used within an organization.

Partner – Shared with specific business partners.

Composite – Combines multiple APIs into one call.

#### DOG CAT API
- you request, the API responds with a image

```python
import numpy as np
import pandas as pd

import requests
from IPython.display import Image, display
```

```python
url = "https://dog.ceo/api/breeds/image/random"

response = requests.get(url)
response.json()
# javascript object notation --- like dictionaries in python

status = response.status_code

if status == 200:
    data = response.json()
    dog_url = data['message']
    print("Here is the image of a dog", dog_url)  
    display(Image(url = dog_url))#, width= 300 , height= 400) 
else:
    print("Failed to fetch data!!")
```

```python
city = "London"
units = "metric"

url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units={units}"

response = requests.get(url)
data = response.json()
print(response.status_code)

# main_weather = data['weather'][0]['main']
# weather_descr = data['weather'][0]['description']

# city Temperature Humidity Pressure
city = data['name']
temp = data['main']['temp']
humidity = data['main']['humidity']
pressure = data['main']['pressure']


weather = {
    'City': [city],
    'Temperature (C)': [temp],
    'Humidity' : [humidity],
    'Pressure': [pressure]
}

df = pd.DataFrame(weather)
df
```
