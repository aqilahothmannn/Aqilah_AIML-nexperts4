- Categories of AI:
1. Narrow AI - Weak - 
    - can perform specified tasks - not beond the scope
    - e.g insta,yt recommendation systems, chatbots

2. General AI - Strong - 
    - possess human like intelligence across domains
    - can do general task, or interconnected tasks
    - e.g booking a ticket, playing chess, writing books, perform surgery just like a human being
    - e.g. - things like agents - are capable of doing multimodal taks [ text, audio, video]
    - 
    
- AGI - Artifical General intelligence - level of AI that matches human intelligence in understanding, learning, and problem solving


3. Super AI - Strong 
    - superintelligent systems, are capable of things beyond human capabilities
    - e.g. Jarvis, ultron, skynet
    - 

- weak and strong are just terms defined compared to human ability
    - neither forms of strong ai exist yet
