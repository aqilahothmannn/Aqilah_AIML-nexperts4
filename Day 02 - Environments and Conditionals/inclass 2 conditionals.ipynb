# $$Conditionals$$

- let your program choose what to do based on a condition
- keywords:
    - `if` - run a block of code ONLY IF a condition is met
    - `elif` - if previous condition is not met, try this one
    - `else` - if all else fails, do this

- 
- end the condition with a **colon** `:`
- give the indent of 4 spaces (is the standard)
- 

- 
- Python is an Indented language
    - indents have meaning
    - indents ---> spaces before lines of code
    - Line by line execution of code
    - 

```python
# simple if clause ---
age = 15

if age >=18:
    print("---Adult---")
```

```python
age = 50

# if else ladder

if age >=70:
    print("---Old Aged---")
elif age >=40:
    print("---Middle Aged---")
elif age >=18:
    print("---Adult---")
elif age>=13:
    print("---Teen---")
elif age>= 4:
    print("---Child---")
elif age >= 2:
    print("---Toddler---")
elif age >= 0:
    print("---Newborn---")

# stop at the first condition met
```

```python
temp = 30

if temp <= 28:
    print("It is cold")
else:
    print("It is hot")
```

### Exercise 2:
- Create a BMI calculator
- Ask input from users and Give them their BMI

Body Mass Index -  BMI = weight (kg) / height (m)²
- height you can ask for feet or inches 
- BMI is commonly classified as follows:
- *Underweight*: BMI less than 18.5
- *Normal* weight: BMI 18.5–24.9
- *Overweight*: BMI 25–29.9
- *Obese*: BMI 30 or higher

```python
# BMI Calculator
 
weight = float(input("Weight (kg): "))
height = float(input("Height (ft): "))
 
# 3.28084ft = 1m
# BMI = kg/m^2
 
BMI = weight / ((height/3.28084)**2)
 
if BMI < 18.5:
    print(f"BMI:{BMI: .2f}, Underweight")
elif 18.5 <= BMI <= 24.9:
    print(f"BMI:{BMI: .2f}, Normal")
elif 25 <= BMI <= 29.9:
    print(f"BMI:{BMI: .2f}, Overweight")
else:
    print(f"BMI:{BMI: .2f}, Obese")
```
