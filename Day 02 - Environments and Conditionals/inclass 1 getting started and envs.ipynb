### Virtual Environments
- right now you downloaded a python version, and used it to execute python code
- 
- now, we will create a space in which we will use a specific version of python to run code
- everytime you create a new project, it is suggested that you create a new virtual environment for it
- 

- Working directory is the `main folder`
-  write in terminal:
    - `python -m venv .venv`
    - first call the application using `python`
    - `-m` ---> flag to use a `module`
    - `venv` ---> is a folder in interpreter that decides how virtual environments look like
    - `.venv` ---> name of the environment --- it can be anything you wish

VS code can let you do this process, without a single line of code!!!
- but you cannot name your environment anything but .venv
- it is what VS code identifies as the virtual environment folder

- the purpose is containerization
- of all the dependencies in one place
