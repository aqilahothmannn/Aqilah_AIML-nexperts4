Example:
- Class: Student (general blueprint that needs - name, grade)
- Objects: Alia, Aathirah, Nazihah ( real students with actual data)

```python
# syntax for creation of class
class Student:
    # when you call a class ---> this function runs
    # it asks for the parameters it has here
    def __init__(self, name, grade, age):
        self.name = name 
        self.grade = grade 
```

```python
# syntax for creation of ---> objecct
student1 = Student("Alia", 9, 24)
```

Example 2: 
- Create a class "Possessions" - everything a person owns
    - save properties like: type, age, color, brand
    - methods / actions 
        - details of the possession
        - action method that tells what the possesion does 
- Create objects for it - "Phone", "Car", 

```python
class Possessions:
    # these are attributes saved in class
    class_name = "possessions"

    # object properties that we can get from a user as well
    def __init__(self, type, age, color, brand, what_action):
        # properties/ attributes ---> data we want to save about our possesions
        self.type = type
        self.age = age
        self.color = color
        self.brand = brand
        self.what_action = what_action

    # Method
    def details(self):
        print(f"I have a {self.brand} {self.type}. \nIt is {self.age} years old.\nIt is {self.color} in color")

    # method
    def action(self):
        print(f"My {self.type} helps me {self.what_action}")
```

```python
phone = Possessions("Phone",3,"Purple","Apple","call and use internet")
computer = Possessions("Computer",5,"Black","HP","work, learn and play")
car = Possessions("car",4, "White", "Toyota", "get to places and drive around")
```

```python
#  access details of an object
phone.type
phone.brand

phone.class_name

phone.details()
computer.action()
```
