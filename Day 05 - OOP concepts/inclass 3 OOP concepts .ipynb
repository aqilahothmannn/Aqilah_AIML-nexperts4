### 1. Encapsulation - 
- Keeping together the data and the methods that work on that data
- Hiding the inner details
- 
- Ḍata should be changed through methods only

````python
class Animal:
    
    def __init__(self, name,age, species):
        self.name = name
        self.species = species
        
        # we want to keep as hidden
        # later on other programmers see this code, they will deduce that
        # Don't change this attribute directly from outside
        self._age = age

    # encapsulation idea is ---> not touching _age directly
    # but using getter and setter methods
    
    # getter methods --- Age
    def get_age(self):
        return self._age
    
    # getter method --- Species
    def get_species(self):
        return self.species

    # setter methods
    def set_age(self, new_age):
        # one could keep checks here as well like, 
        # example: not allowing negative nubmers for age
        if new_age > 0:
            self._age = new_age
        else:
            print("Age cannot be negative")
        
        return f"The new age is {new_age}"
    
    def introduce(self):
        print(f"Hello! I am ---> {self.name}.\n---> I am a {self.species} and I am {self._age} years old.")
````

````python
pet1 = Animal("bruno", 4, "koala")
````

````python
pet1.get_species()
````

````python
pet1.get_age()

pet1.set_age(-2)

pet1.get_age()


# if you are doing

pet1._age = 5 
# this works, but it will break the principle/idea of ENCAPSULATION

# you could change things from outside, but by convention everyone agrees that
# if anything start with _ like _age ---> treated as an internal 
# and you should avoid touching it directly
````

````python
pet1.introduce()
````

### 2. Inheritance
- a new child class that can reuse the code from an existing parent class
- child classes can add new things or change as well
- 
- Let's create these classes for animal: Dog, Cat, Bird
- All animals - 
    - different behaviours like Sounds
    - 
- 

````python
class Dog(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Dog") # super() refers to the parent class ---> Animal
        # 
````

````python
dog1 = Dog("Tommy", 12)
dog2 = Dog("Bruno", 4)
````

````python
dog1.get_species()

dog1.introduce()
````

### 3. Polymorphism
- 

````python
class Cat(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Cat") # super() refers to the parent class ---> Animal
        # 

class Bird(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Bird") # super() refers to the parent class ---> Animal
        # 
````

````python
cat1 = Cat("Noor", 3)
cat2 = Cat("Bruno", 2)
````

````python
cat1.get_species()

cat1.introduce()
````

````python
bird1 = Bird("Shebu", 12)
bird2 = Bird("Mitthu", 4)
````

````python
bird1.get_species()

bird1.introduce()
````
